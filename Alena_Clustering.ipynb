```python
import pandas as pd
df = pd.read_csv('Data_credit.csv',  encoding="windows-1252", header=None)
words = df.loc[:,0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = words
s = ''
for i in range(0, len(words)):
    s += words[i]+ ' '
text = s.split(' ')
vectorizer = CountVectorizer()
vectorizer.fit_transform(text)
v = vectorizer.vocabulary_
print(len(v))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(vocabulary = vectorizer.vocabulary_, max_df = 0.99)

#df = pd.read_csv('Data_credit.csv',  encoding="windows-1252", header=None)
#words = df.loc[:,0]
el = tfidf.fit_transform(words)
```

```python
from sklearn.cluster import KMeans
import numpy as np
clusters = 20
kmeans = KMeans(n_clusters = clusters, random_state = 42).fit(el)

a = []
for i in range(0, len(words)):
    k = int(kmeans.predict(el[i]))
    a.append(k)
for i in range(0,len(a)):
    print(a[i])
matr = []

```

```python
for i in range(0,clusters):
    matr.append(0)
for i in range(0,len(a)):
    matr[a[i]]+=1
print(matr)
    
```
